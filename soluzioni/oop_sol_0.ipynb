### Esercizio 1: Crea una classe Quadrato con le seguenti caratteristiche:

Attibuti:
- lato

Metodi:
- get_lato
- get_diagonale
- get_perimetro
- get_area

```python
class Quadrato():
    def __init__(self, lato):
        self.lato = lato

    def get_lato(self):
        return self.lato
    
    def get_diagonale(self):
        return self.lato * (2 ** 0.5)
    
    def get_perimetro(self):
        return self.lato * 4
    
    def get_area(self):
        return self.lato ** 2
```

```python
# Esempio

q_1 = Quadrato(1)
print("Lato:", q_1.get_lato())
print("Diagonale:", q_1.get_diagonale())
print("Perimetro:", q_1.get_perimetro())
print("Area:", q_1.get_area())
```

### Esercizio 2: Crea una classe Rettangolo con le seguenti caratteristiche:

Attibuti:
- base
- altezza

Metodi:
- get_base
- get_altezza
- get_diagonale
- get_perimetro
- get_area

```python
class Rettangolo():
    def __init__(self, base, altezza):
        self.base = base
        self.altezza = altezza

    def get_base(self):
        return self.base
    
    def get_altezza(self):
        return self.altezza
    
    def get_diagonale(self):
        return (self.base ** 2 + self.altezza ** 2) ** 0.5
    
    def get_perimetro(self):
        return 2 * (self.base + self.altezza)
    
    def get_area(self):
        return self.base * self.altezza
```

```python
# Esempio
r_1 = Rettangolo(2, 1)
print("Base:", r_1.get_base())
print("Altezza:", r_1.get_altezza())
print("Diagonale:", r_1.get_diagonale())
print("Perimetro:", r_1.get_perimetro())
print("Area:", r_1.get_area())
```

### Esercizio 3: Crea una classe triangolo con le seguenti caratteristiche:

Attibuti:
- lato_1
- lato_2
- lato_3
> Memo: un triangolo per essere valido nessun lato deve essere maggiore della somma degli altri due

Metodi:
- get_lato_1
- get_lato_2
- get_lato_3
- get_perimetro
- get_area
> Memo: per calcolare l'area puoi usare la formula di Erone: $Area = \sqrt{p(p-a)(p-b)(p-c)}$ con $p$ il semiperimetro.

```python
class Triangolo():
    def __init__(self, lato_1, lato_2, lato_3):
        if (lato_1 >= lato_2 + lato_3 or
            lato_2 >= lato_1 + lato_3 or
            lato_3 >= lato_1 + lato_2):
            raise ValueError("Triangolo non valido: un lato è >= alla somma degli altri due")
        self.lato_1 = lato_1
        self.lato_2 = lato_2
        self.lato_3 = lato_3

    def get_lato_1(self):
        return self.lato_1
    
    def get_lato_2(self):
        return self.lato_2
    
    def get_lato_3(self):
        return self.lato_3
    
    def get_perimetro(self):
        return self.lato_1 + self.lato_2 + self.lato_3
    
    def get_area(self):
        s = self.get_perimetro() / 2
        return (s * (s - self.lato_1) * (s - self.lato_2) * (s - self.lato_3)) ** 0.5
```

```python
# Esempio
triangolo_1 = Triangolo(3, 4, 5)
print("Lato 1:", triangolo_1.get_lato_1())
print("Lato 2:", triangolo_1.get_lato_2())
print("Lato 3:", triangolo_1.get_lato_3())
print("Perimetro:", triangolo_1.get_perimetro())
print("Area:", triangolo_1.get_area())
```

### Esercizio 4: Crea una classe cerchio con le seguenti caratteristiche:

Attibuti:
- raggio

Metodi:
- get_raggio
- get_diametro
- get_circonferenza
- get_area

```python
import math

class Cerchio():
    def __init__(self, raggio):
        self.raggio = raggio

    def get_raggio(self):
        return self.raggio
    
    def get_diametro(self):
        return self.raggio * 2
    
    def get_circonferenza(self):
        return 2 * math.pi * self.raggio
    
    def get_area(self):
        return math.pi * (self.raggio ** 2)
```

```python
cerchio_1 = Cerchio(1)
print("Raggio:", cerchio_1.get_raggio())
print("Diametro:", cerchio_1.get_diametro())
print("Circonferenza:", cerchio_1.get_circonferenza())
print("Area:", cerchio_1.get_area())        

# per arrotondare a 2 cifre decimali, si può usare la funzione round()
print("\nArrotondamento a 2 cifre decimali:")
print("Circonferenza arrotondata a 2 cifre decimali:", round(cerchio_1.get_circonferenza(), 2))
print("Area arrotondata a 2 cifre decimali:", round(cerchio_1.get_area(), 2))
```
