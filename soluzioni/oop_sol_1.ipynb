### Esercizio 1: Crea una classe `Carrello` con le seguenti caratteristiche:

Attributi:
- `lista_prodotti`: dizionario con le associazioni {prodotto: quantità}
- `lista_prezzi`: dizionario con le associazioni {prodotto: prezzo}

Metodi:
- `get_lista_prodotti`: ritorna il dizionario dei prodotti con le relative quantità
- `get_lista_prezzi`: ritorna il dizionario dei prodotti con i relativi prezzi
- `get_n_prodotti`: ritorna il numero totale di oggetti nel carrello (somma delle quantità)
- `get_costo_totale`: ritorna il costo totale del carrello (somma di quantità × prezzo per ogni prodotto)
- `aggiungi_prodotto(nome, prezzo, quantità)`: aggiunge un prodotto al carrello;
  se il prodotto è già presente, incrementa la quantità
- `rimuovi_prodotto(nome, quantità)`: rimuove una quantità di un prodotto dal carrello;
  se la quantità scende a zero o meno, rimuove il prodotto completamente;
  se il prodotto non esiste, solleva un ValueError

```python
class Carrello():

    def __init__(self):
        self.__lista_prodotti = {}
        self.__lista_prezzi = {}
    
    def get_lista_prodotti(self):
        return self.__lista_prodotti
    
    def get_lista_prezzi(self):
        return self.__lista_prezzi

    def get_n_prodotti(self):
        prodotto_totale = 0
        for prodotto in self.__lista_prodotti:
            prodotto_totale += self.__lista_prodotti[prodotto]
        return prodotto_totale
    
    def aggiungi_prodotto(self, prodotto, quantita, prezzo):
        if prodotto in self.__lista_prodotti:
            self.__lista_prodotti[prodotto] += quantita
        else:
            self.__lista_prodotti[prodotto] = quantita
            self.__lista_prezzi[prodotto] = prezzo
    
    def rimuovi_prodotto(self, prodotto, quantita):
        if prodotto in self.__lista_prodotti:
            if self.__lista_prodotti[prodotto] > quantita:
                self.__lista_prodotti[prodotto] -= quantita
            else:
                del self.__lista_prodotti[prodotto]
                del self.__lista_prezzi[prodotto]
        else:
            raise ValueError
    
    def get_conto(self):
        tot = 0
        for prodotto in self.__lista_prodotti:
            tot += self.__lista_prodotti[prodotto]*self.__lista_prezzi[prodotto]
        return tot
```

```python
class Carrello():

    def __init__(self):
        self.lista_prodotti = {}
        self.lista_prezzi = {}

    def get_lista_prodotti(self):
        return self.lista_prodotti
    
    def get_lista_prezzi(self):
        return self.lista_prezzi
    
    def get_n_prodotti(self):
        tot = 0
        for prodotto in self.lista_prodotti:
            tot += self.lista_prodotti[prodotto]
        return tot
    
    def get_costo_totale(self):
        tot = 0
        for prodotto in self.lista_prodotti:
            tot += self.lista_prodotti[prodotto] * self.lista_prezzi[prodotto]
        return tot
    
    def aggiungi_prodotto(self, nome_prodotto, prezzo_prodotto, quantita_prodotto):
        if nome_prodotto in self.lista_prodotti:
            self.lista_prodotti[nome_prodotto] += quantita_prodotto
        else:
            self.lista_prodotti[nome_prodotto] = quantita_prodotto
            self.lista_prezzi[nome_prodotto] = prezzo_prodotto
    
    def rimuovi_prodotto(self, nome_prodotto, quantita_prodotto):
        if nome_prodotto not in self.lista_prodotti:
            raise ValueError(f"Prodotto '{nome_prodotto}' non presente nel carrello")
    
        if self.lista_prodotti[nome_prodotto] > quantita_prodotto:
            self.lista_prodotti[nome_prodotto] -= quantita_prodotto
        else:
            del self.lista_prodotti[nome_prodotto]
            del self.lista_prezzi[nome_prodotto]
    
```

```python
# Esempio
carrello_1 = Carrello()
carrello_1.aggiungi_prodotto("latte", 1.5, 2)
carrello_1.aggiungi_prodotto("pane", 0.8, 3)
carrello_1.aggiungi_prodotto("uova", 0.2, 12)

prodotti = carrello_1.get_lista_prodotti()
prezzi = carrello_1.get_lista_prezzi()
print("Prodotti nel carrello:", prodotti)
print("Prezzi dei prodotti:", prezzi)
print("Numero totale di prodotti:", carrello_1.get_n_prodotti())
print("Costo totale:", carrello_1.get_costo_totale(), "€")

carrello_1.rimuovi_prodotto("pane", 2)
print("\nDopo aver rimosso 2 unità di pane:")
print("Prodotti nel carrello:", prodotti)
print("Prezzi dei prodotti:", prezzi)
print("Numero totale di prodotti:", carrello_1.get_n_prodotti())
print("Costo totale:", carrello_1.get_costo_totale(), "€")

carrello_1.rimuovi_prodotto("pesce", 2)
```
