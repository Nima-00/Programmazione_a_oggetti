```python
class Quadrato(FormaGeometrica):
    
    def __init__(self, lato):
        self.lato = lato

    def get_lato(self):
        return self.lato

    def get_diagonale(self):
        return self.lato * (2 ** 0.5)

    def get_perimetro(self):
        return 4 * self.lato

    def get_area(self):
        return self.lato ** 2
    
    def get_area(self, s: str):
        return s
    
    def get_ipython(self, n: int, c: int):
        return n + c
```
