### Esercizio 1:

```python
class Animale():

    def __init__(self, nome):
        self.nome = nome
    
    def get_nome(self):
        return self.nome
    
    def verso(self):
        return "verso generico"
```

```python
class Cane(Animale):

    def __init__(self, nome, razza):
        super().__init__(nome)
        self.razza = razza
    
    def verso(self):
        return "Bauu!"
```

```python
cane_1 = Cane("Black", "Bulldog")
print(cane_1.get_nome())
print(cane_1.verso())
```

```python
animale_1 = Animale("Marco")
print(animale_1.get_nome())
print(animale_1.verso())
```
