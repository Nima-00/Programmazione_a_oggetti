```python
class rectangle():
    def __init__(self, width: float, height: float):
        self.__width = width
        self.__height = height

    def get_width(self):
        return self.__width

    def get_height(self):
        return self.__height 

    def get_area(self):
        return self.__width*self.__height
```

```python
R_1 = rectangle(10, 5)
print(R_1.get_area())
print(R_1.get_width())
print(R_1.get_height())
```

```python
class GeometricShape:
    def __init__(self, name):
        self.name = name

    def get_name(self):
        return self.name
```

```python
class Square(GeometricShape):
    def __init__(self, name, side_length):
        super().__init__(name)






        
        self.side_length = side_length

    def get_area(self):
        return self.side_length ** 2
```

```python
S_1 = Square("Quadrato", 4)
print(S_1.get_name())
print(S_1.get_area())
```

```python
### Ereditarietà

class Animale:
    def __init__(self, nome, età):
        self.nome = nome        # attributo
        self.età = età          # attributo

    def descrivi(self):         # metodo
        return f"Sono {self.nome} e ho {self.età} anni"

animale = Animale("Leo", 5)
print(animale.descrivi())   # "Sono Leo e ho 5 anni"
```

```python
class Cane(Animale):            # Cane eredita da Animale
    def __init__(self, nome, età, razza):
        super().__init__(nome, età)  # chiama il costruttore del padre
        self.razza = razza           # attributo aggiuntivo

    def abbaia(self):                # metodo aggiuntivo
        return f"{self.nome} dice: Bau!"

    def descrivi(self):              # sovrascrive il metodo del padre
        return f"Sono {self.nome}, un {self.razza}"

cane = Cane("Rex", 3, "Labrador")
print(cane.abbaia())    # "Rex dice: Bau!"
print(cane.descrivi())  # "Sono Rex, un Labrador"
```

```python
from abc import ABC, abstractmethod

class FormaGeometrica(ABC):

    @abstractmethod
    def get_area(self):
        pass
```

```python
class Triangle(FormaGeometrica):

    def __init__(self, base: float, height: float):
        self.base = base
        self.__height = height
    
    def get_area(self):
        area = self.base * self.__height / 2
        return area
    
    def __perimetro(self):
        p = 2
        return p
    
    def get_perimetro(self):
        return self.__perimetro()
```

```python
A = Triangle(1, 2)
print(A.base)
print(A)
A.get_area()
A.get_perimetro()
```

```python
import math

class Cerchio():

    def __init__(self, radius: float):
        self.radius = radius
    
    def get_perimetro(self, ) -> float:
        return 2*math.pi*self.radius
    
    def get_area(self, ) -> float:
        return math.pi*self.radius**2

```

```python
cerchio = Cerchio(1)
print(cerchio.get_perimetro())
print(cerchio.get_area())

cerchio= Cerchio(2)
print(cerchio.get_perimetro())
print(cerchio.get_area())
```

```python
class Quadrato():
    """Implementazione della forma quadrato"""
    name = "Quadrato"

    def __init__(self, lato: float) -> None:
        """Costruttore"""
        self.__lato = lato
    
    def get_lato(self) -> float:
        """Ritorna il lato"""
        return self.__lato
    
    def get_perimetro(self) -> float:
        """Ritorna il perimetro"""
        return 4*self.__lato
    
    def get_area(self) -> float:
        """Ritorna l'area"""
        return self.__lato**2
    
```

```python
def fattoriale(n):
    if n == 0:
        return 1
    else:
        return n * fattoriale(n-1)
```

```python
def serie(n):
    if n < 0:
        return 0
    else:
        return 1/fattoriale(n) + serie(n-1)
```

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
def fibonacci_iterativo(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
```

```python
for i in range(10):
    print(fibonacci_iterativo(i))
```

```python
for i in range(10):
    print(fibonacci(i))
```
