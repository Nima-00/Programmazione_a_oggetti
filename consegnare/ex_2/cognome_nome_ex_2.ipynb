```python
lista_aule = utils.reaf_file(file_path="input.txt")
```

```python
for aula in lista_aule:
    print(aula)
    for attrezzatura in aula.get_attrezzature():
        print(f"- {attrezzatura}")
    print()
```

```python
for aula in lista_aule:
    if aula.get_edificio() == "Povo2":
        aula.add_attrezzature(utils.Attrezzatura("LavagnaInterattiva", f"LIM-{aula.get_nome()}", utils.Data(16, 1, 2023)))
    
    for attrezzatura in aula.get_attrezzature():
        if attrezzatura.get_data_prossima_revisione().check_data_passata():
            attrezzatura.cambia_data_prossima_revisione(utils.Data(16, 1, 2027))
```
