## Pile

```python
pila = utils.PilaListBased()
# pila = utils.PilaArrayBased()
# pila = utils.PilaLinkedListBased()
```

```python
print(pila, "\t\t<- push(1)")
pila.push(1)
print(pila, "\t\t<- push(2)")
pila.push(2)
print(pila, "\t\t<- push(3)")
pila.push(3)
print(pila)
for i in range(pila.size()):
    print(pila, "\t\t->", pila.pop())
```

## Coda

```python
coda = utils.QueueListBased()
# coda = utils.QueueArrayBased()
# coda = utils.QueueLinkedListBased()
```

```python
print("\t\t", coda, "\t\t<- push(1)")
coda.push(1)
print("\t\t", coda, "\t\t<- push(2)")
coda.push(2)
print("\t\t", coda, "\t<- push(3)")
coda.push(3)
print("\t\t", coda)
for i in range(coda.size()):
    print(coda.pop(), "<-\t\t", coda)
```
